# CLO2/ Develop Python Function for Descriptive Statistics

```python
# Import pandas library
import pandas as pd

# Load the dataset
DS = pd.read_csv("uae_used_cars_10k.csv")

# descriptive Statistics for all columns
print(DS.describe().T)   # T = Transpose

# create function to display descriptive statistics
def Desc_stat(ds, var):
    mean = ds[var].mean()
    median = ds[var].median()
    mode = ds[var].mode()[0]
    minimum = ds[var].min()
    maximum = ds[var].max()
    range_val = maximum - minimum
    std_dev = ds[var].std()
    variance = ds[var].var()
    skewness = ds[var].skew()
    kurtosis = ds[var].kurt()
    count = ds[var].count()

    Quart = [
        # Minimum
        ds[var].quantile(0),
        # 25th percentile
        ds[var].quantile(0.25),
        # Median
        ds[var].quantile(0.50),
        # 75th percentile
        ds[var].quantile(0.75),
        # Maximum
        ds[var].quantile(1),
        # Interquartile range
        ds[var].quantile(0.75) - ds[var].quantile(0.25)
    ]

    summary = {
        "Average": mean,
        "Median": median,
        "Mode": mode,
        "Minimum": minimum,
        "Maximum": maximum,
        "Range": range_val,
        "Std Dev": std_dev,
        "Variance": variance,
        "Skewness": skewness,
        "Kurtosis": kurtosis,
        "Minimum (0%)": Quart[0],
        "25th Percentile": Quart[1],
        "50th Percentile (Median)": Quart[2],
        "75th Percentile": Quart[3],
        "Maximum (100%)": Quart[4],
        "Interquartile Range (IQR)": Quart[5],
        "Count": count
    }

    return summary


```

# CLO2/ Find the Descriptive Statistics for Dependent Variables

```python
# Random sample of size 150
sample = DS.sample(n=150, random_state=1)

# Apply the descriptive statistics function on variables
price_result = Desc_stat(sample, 'Price')
mileage_result = Desc_stat(sample, 'Mileage')
year_result = Desc_stat(sample, 'Year')

# Print results
print("\nDescriptive Statistics for 'Price' (Sample of 150 cars):\n")
for stat, value in price_result.items():
    print(f"{stat}: {value}")

print("\nDescriptive Statistics for 'Mileage' (Sample of 150 cars):\n")
for stat, value in mileage_result.items():
    print(f"{stat}: {value}")

print("\nDescriptive Statistics for 'Year' (Sample of 150 cars):\n")
for stat, value in year_result.items():
    print(f"{stat}: {value}")

```

```python
# Systematic Sampling
k = 66  # Sample interval (every 66th record)
systematic_sample = DS.iloc[::k]  # Picks every 66th record from the dataset

# Descriptive statistics for 'Price'
price_result = Desc_stat(systematic_sample, 'Price')

# Descriptive statistics for 'Mileage'
mileage_result = Desc_stat(systematic_sample, 'Mileage')

# Descriptive statistics for 'Year'
year_result = Desc_stat(systematic_sample, 'Year')

# Print Results
print("\nSystematic Sample (every 66th record) - Descriptive Statistics for 'Price':\n")
for stat, value in price_result.items():
    print(f"{stat}: {value}")

print("\nSystematic Sample (every 66th record) - Descriptive Statistics for 'Mileage':\n")
for stat, value in mileage_result.items():
    print(f"{stat}: {value}")

print("\nSystematic Sample (every 66th record) - Descriptive Statistics for 'Year':\n")
for stat, value in year_result.items():
    print(f"{stat}: {value}")

# Print total records selected
print(f"\nTotal records selected: {len(systematic_sample)}")
```

# CLO2/ Perform Hypothesis Test for Correlation Between Variables

Perform Pearsons Correlation Test between Price and Mileage

```python
# Perform Pearsons Correlation Test between Price and Mileage
from scipy.stats import pearsonr

r, p_value = pearsonr(DS['Price'], DS['Mileage'])
DS_pearson = len(DS) - 2

# Print Results
print("Pearsons Correlation Test Results:")
print(f"1. Pearsons Correlation Coefficient (r): {r:.4f}")
print(f"2. P-Value: {p_value:.4f}")
print(f"3. Degrees of Freedom: {DS_pearson}")

# Interpretation
if p_value < 0.05:
    print("\nConclusion: Since the p-value is less than 0.05, we reject the null hypothesis.")
    print("There is a statistically significant correlation between Price and Mileage.")
else:
    print("\nConclusion: Since the p-value is greater than 0.05, we accept the null hypothesis.")
    print("There is no statistically significant correlation between Price and Mileage.")
```

Perform Spearmans Correlation Test

```python
 # Perform Spearmans Correlation Test
from scipy.stats import spearmanr

r, p_value = spearmanr(DS['Price'], DS['Mileage'])
DS_spearman = len(DS) - 2

# Print the results
print("Spearmans Correlation Test Results:")
print(f"1. Spearmans Correlation Coefficient (r): {r:.4f}")
print(f"2. P-Value: {p_value:.4f}")
print(f"3. Degrees of Freedom: {DS_spearman}")

# Interpretation
if p_value < 0.05:
    print("\nConclusion: Since the p-value is less than 0.05, we reject the null hypothesis.")
    print("There is a statistically significant correlation between Price and Mileage.")
else:
    print("\nConclusion: Since the p-value is greater than 0.05, we accept the null hypothesis.")
    print("There is no statistically significant correlation between Price and Mileage.")
```

# CLO3/ Simple Linear Regression

```python
# Importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Load dataset
data = pd.read_csv("uae_used_cars_10k.csv")

# Select variables
X = data[['Mileage']]     # Independent variable
y = data['Price']         # Dependent variable

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Train the Simple Linear Regression model
regressor = LinearRegression()
regressor.fit(X_train, y_train)

# Predict the test set results
y_pred = regressor.predict(X_test)

# Compare first few predictions with actual values
comparison = pd.DataFrame({'Actual': y_test[:5].values, 'Predicted': y_pred[:5]})
print(comparison)

# Visualize Training Results
plt.scatter(X_train, y_train, color='red')
plt.plot(X_train, regressor.predict(X_train), color='blue')
plt.title('Mileage vs Price (Training Set)')
plt.xlabel('Mileage (km)')
plt.ylabel('Price (AED)')
plt.show()

# Visualize Test Results
plt.scatter(X_test, y_test, color='orange')
plt.plot(X_train, regressor.predict(X_train), color='green')
plt.title('Mileage vs Price (Test Set)')
plt.xlabel('Mileage (km)')
plt.ylabel('Price (AED)')
plt.show()

# Display regression equation and model accuracy
a = regressor.coef_[0]
b = regressor.intercept_
r2 = r2_score(y_test, y_pred)

print("Final Regression Equation: Price = {:.2f} * Mileage + {:.2f}".format(a, b))
print("R Score:", round(r2, 4))
```

# CLO3/ Logistic Regression

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix

# Create a binary column "Affordable" (1 = if price < 100,000 AED, else 0)
data['Affordable'] = np.where(data['Price'] < 100000, 1, 0)

# Define features and target
X = data[['Year', 'Mileage']]
y = data['Affordable']

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Train the Logistic Regression model
log_reg = LogisticRegression(max_iter=1000)
log_reg.fit(X_train, y_train)

# Predict and evaluate
y_pred = log_reg.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", cm)
print("Accuracy:", round(accuracy, 3))
```

# CLO3/ K-Nearest Neighbors (KNN)

```python
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split data
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=0)

# Train KNN model
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)

# Predict and evaluate
y_pred = knn.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", cm)
print("Accuracy:", round(accuracy, 3))
```

# CLO3/ Nave Bayes (NB)

```python
from sklearn.naive_bayes import GaussianNB

# Train and predict using Naive Bayes
nb = GaussianNB()
nb.fit(X_train, y_train)
y_pred = nb.predict(X_test)

# Evaluate performance
accuracy = accuracy_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", cm)
print("Accuracy:", round(accuracy, 3))
```

# CLO3/ Decision Tree (DT)

```python
from sklearn.tree import DecisionTreeClassifier

# Train and predict using Decision Tree
dt = DecisionTreeClassifier(criterion='entropy', random_state=0)
dt.fit(X_train, y_train)
y_pred = dt.predict(X_test)

# Evaluate performance
accuracy = accuracy_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", cm)
print("Accuracy:", round(accuracy, 3))
```
